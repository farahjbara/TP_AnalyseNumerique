### <span style="color:red">  Introduction </span>

L’objet essentiel de ce TP est l’approximation des racines d’une fonction réelle
d’une variable réelle, c’est-à-dire la résolution approchée du problème suivant :
Étant donné une équation non linéaire, à une seul variable, est définie par :
<br>
<br>
 <center > $f(x)=0$  </center>
<br>
La valeur de la variable $X$ qui vérifie cette égalité est appelée solution (ou racine) de
l'équation, elle est notée $X0$. 
Dans beaucoup de cas, on doit recourir aux méthodes numériques car la solution ne peut pas être déterminée analytiquement.
<br>
<br> 
Ainsi, résoudre cette équation revient à définir une succession d'éléments (solutions intermédiaires $X0$ ;$k$ )
qui convergent vers la solution désirée lorsque $K$ tant vers l’infini.
<br>
On ne parle pas de la solution exacte, car souvent elle n'existe pas, cependant on cherche une solution qui soit
approchée de la solution exacte avec une certaine précision.
<br>
Il existe plusieurs techniques permettant de résoudre l'équation non linéaire, ces
méthodes se distinguent par leurs principes et leurs vitesses de convergence. 
<br>
Nous introduisons dans cette section les méthodes de dichotomie (ou de bissection), point fixe
et de Newton. Nous les présentons dans l’ordre de complexité croissante des algorithmes.
<br>

#### Solution théorique (Version python )

```python
def horner(P,x):
    n=len(P)
    valeur=P[n-1]
    for i in range(n-2,-1,-1):
        valeur=valeur*x+P[i]
    return valeur
```

#### Principe de Methode Dichotomie :
La méthode de dichotomie est une méthode pour trouver une solution approchée à une équation $f(x)=0$ . Précisément, supposons que la fonction f est continue sur l'intervalle [a,b], avec f(a)≤0et f(b)≥0 On sait donc qu'il existe au moins un réel c dans l'intervalle [a,b] tel que f(c)=0 L'idée est alors d'évaluer ce que vaut f au milieu de [a,b] et de distinguer les deux cas suivants :
<br>
si $f(a+b/2)≤0$, alors on sait qu'on a une racine dans l'intervalle [a+b/2,b]

sinon, 
<br>
$f(a+b/2)>0 $ et on sait qu'on a une racine dans l'intervalle [a,a+b/2] Ainsi, dans les deux cas, on a trouvé un intervalle de longueur moitié dans lequel est située une racine de l'équation $f(x)=0$
<br>
On recommence alors avec cet intervalle, et ainsi de suite jusqu'à ce qu'on trouve une approximation qui nous convienne.

2.   Le programme dicho.m permette de trouver les solutions de l'équation $f(x)=0$ avec $eps=0.001$.

```python
def dicho(a,b,f,e):
    '''version tant que '''
    
    while (b-a)>e:
        m=(a+b)/2
        if f(m)==0:
            return m
        if f(a)*f(m)<0 :
            b=m
        else :
            a=m
    return m
```

```python
f=lambda x: x**3+x**2-3*x-3
dicho(-2,-1.5,f,0.001)
```

#### 1)

```python
def newton(f,fprime,a,eps):
    x1=a
    x2=x1-(f(x1)/fprime(x1))
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-(f(x1)/fprime(x1))
        m=x2
    return m
    
```

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newton(f,fprime,0,0.001)
```

#### Commentaires :

la méthode de newton n'est qu'une cas particulier de la méthode de point fixe

#### Conclusion Générale

<b> (i) </b> Pour comparer les algorithmes, il est important de tenir compte de la présence ou non des facteurs suivants :
<ol>
<li>assurance de la convergence </li>
<li>vitesse de la convergence </li>
<li>stabilité de l'algorithme /précision des résultats </li>
<li>effcacité des calculs (ex : nombre de fonctions à calculer à chaque itération). </li>
  </ol>
<b>(ii) </b> La méthode de point fixe a surtout un interêt théorique : son analyse mathématique est relativement aisée et nous a permis d'en déduire celle de la méthode de Newton qui en est en fait un cas particulier.

<b>(iii)</b> La méthode de Newton a généralement la faveur des utilisateurs : elle est effectivement très rapide et est certainement à conseiller lorsque le calcul de f' est aisé et lorsque des remarques simples permettent de choisir à coup sûr la donnée initiale dans le domaine de convergence. Pour cela, elle peut être, par exemple, précédée d'une application de quelques itérations de la méthode de dichotomie.
